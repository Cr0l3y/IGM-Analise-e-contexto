Dados correspondentes ao IGM onde iremos analisar as estruturas de linhas, amostra das informações, sendo analisadas dados de estado, gestão, violencia entre outros




```python
#Primeiro importamos o pandas para poder rodar nosso DataFrame

import pandas as pd
```

```python
#Carregamos o arquivo CSV e transformar em DataFrame do Pandas.
#Apos conferimos o arquivo vamos retirar a linha superior onde que poderar
#ocasionar em erros assim conferimos as linhas inferiores invalidas para analise.
# retiramos a linha invalida com o comando "skiprows = NUMERO DE LINHAS"

df = pd.read_excel('dados_IGM-CFA_2017.xlsx',skiprows=1)


#Mostrando as primeiras 5 linhas do DataFrame, onde pode-se restringir a quiandidade colocando o valor nos parenteses()

df.head()
```

```python
# Mostrando as 5 ultimas linhas, sendo que nos continuamente podemos restrigir a quantidade 
df.tail()
```

```python
# Tipos de dados a serem vistos/analisados
df.dtypes
```

```python
#Quantidades de dados em relação a (linhas, colunas)
df.shape
```

```python
# Análise descritiva das colunas numericas
df.describe()
```

```python
#Analise das de texto ultilizando o describe
df.describe(include=[object])
```

```python
#Verificando a quantidade de município classificada pelo seu porte.

df['Porte do Município'].value_counts()
```

```python
# Grafico onde esta os dados classificados
#informando a visualização do grafico no jupyter

%matplotlib inline

# Grafico de barras para visualização dos portes dos muninicipios

df['Porte do Município'].value_counts().plot.bar()
```

```python
# Maior gasto registrado pelo PIB em 2013 
df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'].max()
```

```python
#visualização pela faixa escolhida, que é o maior gasto


df[df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'] == df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'].max()]
```

```python
# Menor gasto registrado pelo PIB em 2013 
df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'].min()
```

```python
#visualização pela faixa escolhida, que é o menor gasto


df[df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'] == df['PIB 2013 - a preços correntes (1000 R$) (IBGE)'].min()]
```

```python
#Estado onde a analise esta sendo feita

df['Estado'].value_counts()
```

```python
#Os tipos de Graus de status em qualidade de Gestão

df['Status em Qualidade da Gestão'].unique()
```

```python
#Quantidade numerica dos status em cada grau de gestão

df['Status em Qualidade da Gestão'].value_counts()
```

```python
# Grafico de dados classificados de forma de pizza para visualização mais ampla sendo informados,
#"outiler", "Com dados" e "sem dados" na nossa base de Status em Qualidade da Gestão.


%matplotlib inline

# Grafico de barras para visualização dos portes dos muninicipios

df['Status em Qualidade da Gestão'].value_counts().plot.pie()
```
